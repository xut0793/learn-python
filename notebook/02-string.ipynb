# 字符串

字符串可以单引号、双引号、三引号。
单引号：''
双引号：""
三引号：""" """  或者 ''' '''

其中三引号可以是单引号或双引号组成，主要用于多行或者换行文本显示。

```python
a = '123'
b = "123"
c = '''
    1
    2
    3
    '''
d = """
    4
    5
    6
    """

print(a)
print(b)
print(c)
print(d)
```

## 在字符串中插入变量的值 f-string

语法要求：`f"{变量名}"`

```python
var = 124
print(f"数字是{var}")
```
